```python
def frac(x):
    return x - math.floor(x)


import numpy as np
from numpy import *
import math

golden = (np.sqrt(5)+1)/2
#Tribonacci constant
tau = 1/3 *(1 + (19-3*np.sqrt(33))**(1/3) + (19+3*np.sqrt(33))**(1/3))
one = np.pi / 3.1416

```

```python
def getErrorMaxLists(num): # sequence = n * num
    errorMaxs = []
    alphaList = np.linspace(0, 1, 100)

    for alpha in alphaList:
        size = 50000
        counts = [] #countDict[n] = ct of i*num < alpha for  1<=i<=n
        errors = []

        currCt = 0
        for n in range(size):
            if frac(n * num) < alpha:
                currCt += 1;
            counts.append(currCt)
    #     print("countDict Done!")
        for n in range(size):
            errors.append(counts[n] - alpha*n)

        errorMaxs.append(max(errors))
    #     print("alpha = ", alpha, " done")
    print("num = ", num, " done!")
    return alphaList, errorMaxs
```

```python
import matplotlib.pyplot as plt

for i in [2,3,5,6,7,8,10,11]:
    print("b = Sqrt[i] = Sqrt of ", i)
    num = np.sqrt(i)
    alphaList, errorMaxs = getErrorMaxLists(num)
#     for i in range(len(alphaList)):
#         print("alpha = %.4f, max error =" % (alphaList[i]), errorMaxs[i])
    plt.scatter(alphaList, errorMaxs)
    plt.xlabel('alpha')
    plt.ylabel('Max Error D')

    plt.show()

```
